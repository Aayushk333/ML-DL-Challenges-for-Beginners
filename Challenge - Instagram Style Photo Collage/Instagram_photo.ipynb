```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
import pandas as pd
import os

os.chdir('../Desktop/ImageCollage dataset')

img1 = cv2.imread('top_left.jpg')
img2 = cv2.imread('top_right.jpg')
img3 = cv2.imread('bottom_left.jpg')
img4 = cv2.imread('bottom_right.jpg')
img5 = cv2.imread('center.jpeg')


rgbim1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
rgbim2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
rgbim3=cv2.cvtColor(img3,cv2.COLOR_BGR2RGB)
rgbim4=cv2.cvtColor(img4,cv2.COLOR_BGR2RGB)
rgbim5=cv2.cvtColor(img5,cv2.COLOR_BGR2RGB)




rgbim1rs = cv2.resize(rgbim1,(200,200))
rgbim2rs = cv2.resize(rgbim2,(200,200))
rgbim3rs = cv2.resize(rgbim3,(200,200))
rgbim4rs = cv2.resize(rgbim4,(200,200))
rgbim5rs = cv2.resize(rgbim5,(100,100))

final=np.zeros((430,430,3),int)
final[10:210,10:210]=rgbim1rs
final[10:210,220:420]=rgbim2rs
final[220:420,10:210]=rgbim3rs
final[220:420,220:420]=rgbim4rs
final[155:275,155:275] =0 #creating black canvas for center image
final[165:265,165:265]=rgbim5rs


#print(final.shape)
updated=final.reshape(184900,3)
np.savetxt('test1.csv',updated,header='r,g,b')
#pd.DataFrame(data = updated).to_csv("submission.csv",index = False, header = ['r','g','b'])
```
